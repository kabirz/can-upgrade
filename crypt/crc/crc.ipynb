```python
import crcmod
import zlib
```

```python
crc32 = crcmod.mkCrcFun(0x104C11DB7, initCrc=~0, rev=True, xorOut=0xFFFFFFFF)
crc32_fun = crcmod.predefined.mkCrcFun('32')
```

```python
data = b'123456789'
hex(crc32(data, crc=0)), hex(zlib.crc32(data)&0xFFFFFFFF), hex(crc32_fun(data))
```

```python
crc32_p = crcmod.predefined.Crc('32')
crc32_p.update(data)
hex(crc32_p.crcValue)
with open('crc32.h', 'w') as f:
    crc32_p.generateCode('crc32_calc', f, dataType='unsigned char', crcType='unsigned int')
```

```python
import crcmod

# 对齐 CRC-32 (IEEE 802.3) 参数（与 crccalc.com 的 CRC-32 一致）
crc32_func = crcmod.mkCrcFun(
    poly=0x104C11DB7,
    initCrc=0xFFFFFFFF,
    rev=True,
    xorOut=0xFFFFFFFF
)

data = b"123456789"
print(hex(crc32_func(data)))  # 输出 0xCBF43926
```

```python
modbus_crc16 = crcmod.mkCrcFun(
    poly=0x18005,        # 多项式
    initCrc=0xFFFF,     # 初始值
    rev=True,           # 输入和输出均反转（RefIn + RefOut）
    xorOut=0x0000       # 不异或输出
)

hex(modbus_crc16(b'abcd'))
```

```python
crc16_modbus = crcmod.predefined.Crc('modbus')
with open('modbus_crc.h', 'w') as f:
    crc16_modbus.generateCode('crc_modbus', f, dataType='unsigned char', crcType='unsigned short')

crc16_modbus.update(b'abcd')
hex(crc16_modbus.crcValue)
```

```python

crc16 = crcmod.mkCrcFun(
    poly=0x11021,       # 多项式
    initCrc=0xFFFF,     # 初始值
    rev=0,              # 输入和输出均反转（RefIn + RefOut）
    xorOut=0x0000       # 不异或输出
)
hex(crc16(b'Hello'))
```

```python
from wcwidth import wcwidth

def draw_box(text):
    text_width = sum(wcwidth(c) for c in text)
    box_width = text_width + 2

    print("+" + "-" * box_width + "+")
    print("| " + text + " |")
    print("+" + "-" * box_width + "+")

draw_box("Hello 中文")
```
